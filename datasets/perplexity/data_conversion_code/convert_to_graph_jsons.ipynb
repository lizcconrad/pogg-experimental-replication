# Convert to Graph JSONs

This notebook converts JSONs that store the "raw" perplexity data obtained from a prolog queries to the `graph_json` format that POGG requires:

```json
{
    'nodes': {
        "idGlowingGrass1": {
            node_properties: {
                node_type: "entity",
                lexicon_key: "idGlowingGrass"
            }
        },
        "green": {
            node_properties: {
                node_type: "property"
            }
        },
    },
    'edges': [
        {
            "edge_name": "idGlowingGrass1_prop_idColor",
            "parent_node": "idGlowingGrass1",
            "child_node": "green",
            "edge_properties": {
                "edge_type": "property",
                "lexicon_key": "idColor"
            }
        },
        {
            "edge_name": "isTouching",
            "parent_node": "idGlowingGrass1",
            "child_node": "idGlowingArea1",
            "edge_properties": {
                "edge_type": "relationship"
            }
        }
    ]
}
```

For every node and edge, you can specify a `lexicon_key` if it differs from the node name itself. If not specified, POGG will assume the node name as the key.

You may also specify "properties" which will be read in and stored as node/edge properties on the NetworkX graph elements.


````python
import re
import os
import pathlib
import json
from perplexity_original.data_conversion_code.perplexity_data_handling import convert_data_instance_to_json

mode = "test"


data_directory = f"/Users/lizcconrad/Documents/PhD/POGG/data_handling/perplexity/{mode}"

# Iterate over files in directory
game_names = sorted([name for name in os.listdir(data_directory)])
# print(game_names)
game_names = ["AtomicCity"]

for game_name in game_names:
    if os.path.isdir(os.path.join(data_directory, game_name)):
        raw_data_directory = os.path.join(data_directory, game_name, "data_jsons")
        raw_data_json_path = os.path.join(raw_data_directory, f"{game_name}_raw_data.json")

        # mkdirs
        graph_directory = os.path.join(data_directory, game_name, "graphs")
        entity_graphs_dir = pathlib.Path(graph_directory, "entity_graphs")
        pathlib.Path(graph_directory).mkdir(parents=True, exist_ok=True)
        pathlib.Path(entity_graphs_dir).mkdir(parents=True, exist_ok=True)

        full_graph_json = {
            'nodes': {},
            'edges': []
        }

        # write graph_json for just one entity in the "game world"
        with open(os.path.join(raw_data_directory, raw_data_json_path), "r") as raw_data_file:
            data = json.load(raw_data_file)

            for entry in data:
                single_graph_json = convert_data_instance_to_json(entry)
                single_graph_json_file_name = f"{entry['Entity']}_{game_name}_graph_format.json"
                with open(os.path.join(entity_graphs_dir, single_graph_json_file_name), "w") as entity_json_file:
                    json.dump(single_graph_json, entity_json_file, indent=4)
                full_graph_json['nodes'].update(single_graph_json['nodes'])
                full_graph_json['edges'].extend(single_graph_json['edges'])

        print(full_graph_json)
        # write graph_json for whole "game world"
        with open(os.path.join(graph_directory, f"{game_name}_full_graph_format.json"), "w") as graph_json_file:
            json.dump(full_graph_json, graph_json_file, indent=4)

#
# # for each "game world"...
# for raw_data_json in raw_data_files:
#     if raw_data_json.endswith(".json"):
#         print(raw_data_json)
#         game_name = re.match(r'(.*)_raw_data.json', raw_data_json).group(1)
#
#
#
#         # write graph_json for just one entity in the "game world"
#         with open(os.path.join(raw_data_directory, raw_data_json), "r") as raw_data_file:
#             data = json.load(raw_data_file)
#
#             for entry in data:
#                 single_graph_json = convert_data_instance_to_json(entry)
#                 single_graph_json_file_name = f"{entry['Entity']}_{game_name}_graph_format.json"
#                 with open(os.path.join(entity_graphs_dir, single_graph_json_file_name), "w") as entity_json_file:
#                     json.dump(single_graph_json, entity_json_file, indent=4)
#                 full_graph_json['nodes'].update(single_graph_json['nodes'])
#                 full_graph_json['edges'].extend(single_graph_json['edges'])
#
#         # write graph_json for whole "game world"
#         with open(os.path.join(scenario_graph_dir, f"{game_name}_full_graph_format.json"), "w") as graph_json_file:
#             json.dump(full_graph_json, graph_json_file, indent=4)
````
